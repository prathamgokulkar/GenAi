## Bag-of-words with N-grams

```python
import pandas as pd
```

```python
dataset = pd.read_csv(r".\spam.csv")
```

```python
dataset
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
stopwords = nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
snowballStemmer = SnowballStemmer('english')
```

```python
corpus = []
for i in range(len(dataset)):
    review = re.sub('[^a-zA-Z]', " ", dataset['Message'][i])
    review = review.lower()
    review = [snowballStemmer.stem(word) for word in review if word not in stopwords.words('english')]
review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 500, binary=True, ngram_range=(1,3))
```

```python
X = cv.fit_transform(corpus).toarray()
```
